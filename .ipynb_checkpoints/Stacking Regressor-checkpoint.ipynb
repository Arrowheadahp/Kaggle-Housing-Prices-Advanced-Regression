```python
loc_train = 'data\\train.csv'
loc_test = 'data\\test.csv'

target = 'SalePrice'
evaluation = 'neg_mean_squared_log_error'

raw_train_data = pd.read_csv(loc_train)
y = raw_train_data[target]
raw_train_data = raw_train_data.drop(target, axis=1)
raw_test_data = pd.read_csv(loc_test)
raw_train_data.head(5)
```

```python
def prepare(raw_DF):
    DF = raw_DF.copy()
    
    DF = DF.drop('Id', axis= 1)
    
    return DF
```

```python
train_X = prepare(raw_train_data)
test_X = prepare(raw_test_data)

F = train_X.copy()
F['y'] = y
F.head()
```

```python
y = np.log(y)
```

```python
GBR = GradientBoostingRegressor(n_estimators=8000, 
                               loss= 'huber')
```

```python
# https://scikit-learn.org/stable/supervised_learning.html

Regressors_small = [
    GradientBoostingRegressor(),
    KNeighborsRegressor(), 
    RandomForestRegressor(),
    ExtraTreesRegressor(),
]

Regressors_medium = [ 
    SVR(), 
    LinearSVR(), 
    NuSVR(),
    linear_model.RidgeCV(),
    linear_model.LarsCV(),
    linear_model.LassoCV(),
    linear_model.LassoLarsCV(),
    linear_model.PassiveAggressiveRegressor(), 
]
```

```python
# parameter tuning
gbt_parameters = {
    'n_estimators': [100, 300, 500, 700], 
    'loss': ['ls', 'lad', 'huber'], 
    'warm_start': [True]
}

knn_parameters = {
    'n_neighbors': range(7, 12),
    'weights': ['distance']
}

rf_parameters = {
    'n_estimators': [100, 300, 500, 700], 
    'warm_start': [True]
}

et_parameters = {
    'n_estimators': [100, 300, 500, 700], 
    'warm_start': [True]
}
```

```python
gbt_models = [GradientBoostingRegressor(**params) for params in ParameterGrid(gbt_parameters)]
knn_models = [KNeighborsRegressor(**params) for params in ParameterGrid(knn_parameters)]
et_models = [ExtraTreesRegressor(**params) for params in ParameterGrid(et_parameters)]
rf_models = [RandomForestRegressor(**params) for params in ParameterGrid(rf_parameters)]
```

```python
MLA = Regressors_small
MLA = gbt_models + knn_models + et_models + rf_models
MLA = MLA + Regressors_medium
len(MLA)
```

```python
# List of the Classifiers to be stacked

Stack_estimators = [
    (a.__class__.__name__+str(i), a) for i, a in enumerate(MLA)
]
Stack_estimators[:5]
```

```python
voter = VotingRegressor(Stack_estimators)
voter.fit(train_X, y)
```

```python
# mla = [(f'gbr{i}', GradientBoostingRegressor(loss=random.choice(['huber', 'ls'])))
#                                                           for i in range(100)]
# mla[:5]
```

```python
Meta_models = [
    linear_model.RidgeCV(),
    linear_model.LassoCV(),
    linear_model.LassoLarsCV(), 
    linear_model.ElasticNetCV()
]
names = ['ridge', 'lasso', 'lassolars', 'elasticnet']
stacker = [StackingRegressor(Stack_estimators, final_estimator=mm) for mm in Meta_models]
```

```python
voter_pred = np.exp(voter.predict(test_X))
name = 'Voter'
output = pd.DataFrame({'Id': raw_test_data.Id, target: voter_pred})
output.to_csv(f'preds\\GBR\\{name}stack_7.csv', index = False)
output.head()
```
